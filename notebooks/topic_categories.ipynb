```python
def clean_text(corpus):
    #Remove only numbers
    corpusEsp = [re.sub(r"\[[0-9]+(, [0-9]+)*(,[0-9]+)*\]", "", doc, flags=re.MULTILINE) for doc in corpus]
    corpusEsp = [re.sub(r"[0-9]+", "", doc, flags=re.MULTILINE) for doc in corpusEsp]

    # Convert to lowercase
    corpusEsp = [doc.lower() for doc in corpusEsp]

    # Remove new line characters
    corpusEsp = [re.sub('\s+', ' ', doc) for doc in corpusEsp]

    # Remove distracting single quotes
    corpusEsp = [re.sub("\'", "", doc) for doc in corpusEsp]

  # Remove _
    corpusEsp = [re.sub("_", "", doc) for doc in corpusEsp]

    # Replace punctuation marks
    corpusEsp = [re.sub(r"[^\w\s]", "", doc) for doc in corpusEsp]

    return corpusEsp
```

```python
from gensim.models.keyedvectors import KeyedVectors
import gensim

# Cargar los embeddings GloVe manualmente
def load_glove_embeddings(glove_file):
    embeddings_index = {}
    with open(glove_file, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            coefs = np.asarray(values[1:], dtype='float32')
            embeddings_index[word] = coefs
    return embeddings_index

# Convertir documentos en vectores usando GloVe
def get_document_embeddings(docs, embeddings_index, embedding_dim=300):
    doc_embeddings = []
    for doc in docs:
        words = doc.split()
        word_embeddings = [embeddings_index.get(word, np.zeros(embedding_dim)) for word in words]
        if len(word_embeddings) == 0:
            doc_embedding = np.zeros(embedding_dim)
        else:
            doc_embedding = np.mean(word_embeddings, axis=0)
        doc_embeddings.append(doc_embedding)
    return np.array(doc_embeddings)
    
# Ruta del archivo GloVe (asegúrate de que este archivo esté descargado en tu máquina)
glove_file = '/data/alejandrobuitrago/glove.42B.300d.txt'

# Cargar embeddings GloVe
embeddings_index = load_glove_embeddings(glove_file)

# Obtener los embeddings para los documentos
embedding_dim = 300  # Dimensión de los embeddings GloVe que estás usando
embeddings = get_document_embeddings(docs, embeddings_index, embedding_dim)
```

```python
# Function to get FastText embeddings
def get_fasttext_embeddings(docs, model):
    embeddings = []
    for doc in docs:
        words = doc.split()
        word_embeddings = [model.get_word_vector(word) for word in words]
        doc_embedding = np.mean(word_embeddings, axis=0)
        embeddings.append(doc_embedding)
    return np.array(embeddings)

# Get embeddings for the documents using FastText
embeddings = get_fasttext_embeddings(docs, ft_model)
```
